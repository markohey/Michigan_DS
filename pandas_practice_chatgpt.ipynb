```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6],
    "C": [7, 8, 9]
}, 
index=["Row1", "Row2", "Row3"]
)

print(df)
print(df.dtypes)
print(df.index)
print(df.columns)
```

```python
print("iloc first row:")
print(df.iloc[0])

print("iloc second row, column B:")
print(df.iloc[1, 1])

print("loc row 3:")
print(df.loc["Row3"])

print("loc column A for all rows:")
print(df.loc[:,"A"])
```

```python
mask = (df["A"] > 1) & (df["B"]>4)
print(mask)

filtered = df[mask]
print(filtered)
```

```python
df["D"] = df["A"] + df["B"]
df["E"] = df["A"] * df["B"]

print(df)
```

```python
df = pd.DataFrame({
    "Category": ["A", "B", "A", "B", "A"],
    "Value": [5, 10, 15, 20, 25]
})

df
```

```python
grouped = df.groupby("Category")
result = grouped.agg({"Value": ["mean", "sum"]})
result
```
