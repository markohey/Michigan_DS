## Array Operations

```python
a = np.array([10, 20, 30, 40])
b = np.array([1, 2, 3, 4])

c = a - b
print(c)

d = a*b
print(d)
```

```python
fahrenheit = np.array([0, -10, -5, -15, 0])

celsius = (fahrenheit - 31) * 5/9
celsius
```

```python
celsius > -20
```

```python
celsius % 2 == 0
```

```python
A = np.array([[1, 1], [0, 1]])
B = np.array([[2, 0], [3, 4]])
print(A)
print(B)
print(A*B)
print(A@B)
```

```python
A.shape
```

```python
array1 = np.array([[1, 2, 3], [4, 5, 6]])
print(array1.dtype)

array2 = np.array([[7.1, 8.2, 9.1], [10.4, 11.2, 12.3]])
print(array2.dtype)
```

```python
array3 = array1 + array2
print(array3)
print(array3.dtype)
```

```python
print(array3.sum())
print(array3.max())
print(array3.min())
print(array3.mean())
```
