```python
import pandas as pd

df = pd.DataFrame({
    "A": [1, 2, 3],
    "B": [4, 5, 6],
    "C": [7, 8, 9]
}, 
index=["Row1", "Row2", "Row3"]
)

print(df)
print(df.dtypes)
print(df.index)
print(df.columns)
```

```python
print("iloc first row:")
print(df.iloc[0])

print("iloc second row, column B:")
print(df.iloc[1, 1])

print("loc row 3:")
print(df.loc["Row3"])

print("loc column A for all rows:")
print(df.loc[:,"A"])
```

```python
mask = (df["A"] > 1) & (df["B"]>4)
print(mask)

filtered = df[mask]
print(filtered)
```

```python
df["D"] = df["A"] + df["B"]
df["E"] = df["A"] * df["B"]

print(df)
```

```python
df = pd.DataFrame({
    "Category": ["A", "B", "A", "B", "A"],
    "Value": [5, 10, 15, 20, 25]
})

df
```

```python
grouped = df.groupby("Category")
result = grouped.agg({"Value": ["mean", "sum"]})
result
```

```python
import pandas as pd

df = pd.read_csv("Admission_Predict.csv", index_col=0)
stripped_cols = [x.lower().strip() for x in df.columns]
df.columns = stripped_cols
df.head()
```

```python
admit_mask = df['chance of admit'] > 0.7
admit_mask.head()
```

```python
df.where(admit_mask).head()

# or just use
# df = df[admit_mask]
```

```python
df.where(admit_mask).dropna().head()
```

```python
# Way better method to include where() and dropna()

df[df['chance of admit'] > 0.7].head()
```

```python
df['gre score'].head()
```

```python
df[["gre score", "toefl score"]].head()
```

```python
# multiple boolean mask criteria using | and &

mask = (df['chance of admit'] > 0.7) & (df['chance of admit'] < 0.9)
df[mask]
```

```python
# also works with gt() - greater than and lt() - lower than

df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)
```

```python
# or by chaining them together

df['chance of admit'].gt(0.7).lt(0.9)
```
